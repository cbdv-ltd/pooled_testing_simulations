```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ar_m_e = pd.read_csv("arsenic_ex.csv")
ar_s = pd.read_csv("arsenic_single.csv")
```

```python
ar_m_e["percentage"] = (1 - ar_m_e["percentage"])*100
ar_s["tests"] = (1 - ar_s["tests"])*100
ar_s.head()
```

```python
ar = ar_s.loc[:,["pool", "tests"]].copy()
ar.columns = ["pool", "single"]
arc = pd.merge(left = ar, right = ar_m_e.loc[:,["pool", "percentage"]].copy(),how="outer", on="pool")
arc.columns = ["pool", "single", "matrix"]
arc
```

```python
# sb.scatterplot(data=arc, x="pool", y=["single", "matrix"])
sb.scatterplot(data=arc, x="pool", y="matrix")
sb.scatterplot(data=arc, x="pool", y="single", marker ="+")

plt.show()
```

```python
ar_m_e = pd.read_csv("as_matrix.csv")
ar_s = pd.read_csv("as_single.csv")
ar_m_e["percentage"] = (1 - ar_m_e["percentage"])*100
ar_s["tests"] = (1 - ar_s["tests"])*100
ar = ar_s.loc[:,["pool", "tests"]].copy()
ar.columns = ["pool", "single"]
arc = pd.merge(left = ar, right = ar_m_e.loc[:,["pool", "percentage"]].copy(),how="outer", on="pool")
arc.columns = ["pool", "single", "matrix"]
arcc=arc.copy()
arcc = arcc.set_index("pool")
g = sb.scatterplot(data=arcc, legend=True)
g.set(ylim=[0,90])
plt.title("Arsenic Pool Size Efficency")
plt.xlabel("Pool Size")
plt.ylabel("Tests Saved per 100 Samples")
plt.savefig("arsenic_pool.png")
plt.show()
```
