```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wash_ar_m = pd.read_csv("wash_as_matrix.csv")
wash_ar_s = pd.read_csv("wash_as_single.csv")
wash_ar_m["percentage"] = (1 - wash_ar_m["percentage"])*100
wash_ar_s["tests"] = (1 - wash_ar_s["tests"])*100
cali_ar_m = pd.read_csv("cali_as_matrix.csv")
cali_ar_s = pd.read_csv("cali_as_single.csv")
cali_ar_m["percentage"] = (1 - cali_ar_m["percentage"])*100
cali_ar_s["tests"] = (1 - cali_ar_s["tests"])*100
cali_ar = cali_ar_s.loc[:,["pool", "tests"]].copy()
cali_ar.columns = ["pool", "single"]
arc_c = pd.merge(left = cali_ar, right = cali_ar_m.loc[:,["pool", "percentage"]].copy(),how="outer", on="pool")
arc_c.columns = ["pool", "single_c", "matrix_c"]
wash_ar = wash_ar_s.loc[:,["pool", "tests"]].copy()
wash_ar.columns = ["pool", "single"]
arc = pd.merge(left = wash_ar, right = wash_ar_m.loc[:,["pool", "percentage"]].copy(),how="outer", on="pool")
arc.columns = ["pool", "single_w", "matrix_w"]
arc = pd.merge(left=arc, right = arc_c, how="outer", on="pool")
arcc=arc.copy()
arcc = arcc.set_index("pool")
g = sb.scatterplot(data=arcc, legend=True)
g.set(ylim=[0,90])
plt.title("Arsenic Pool Size Efficency")
plt.xlabel("Pool Size")
plt.ylabel("Tests Saved per 100 Samples")
plt.savefig("arsenic_pool.png")
plt.show()
```
